```python
# definindo quantidade de verdade que sai Cara ou Coroa

import numpy as np

def CoinFlipper(N):
  HeadsNumber = list()
  TailsNumber = list()
  FlippedCoin = np.random.randint(0,2,size=N)
  for i in range(FlippedCoin.size):
    if FlippedCoin[i] == 0:
      HeadsNumber.append(1)
    else:
      TailsNumber.append(1)

  print(len(HeadsNumber)/N)
  print(len(TailsNumber)/N)

CoinFlipper(5000)
```

```python
# create a dot product function
# function checks that both inputs are numpy arrays (isinstance())
# function checks for same length of two input vectors
# returns dot product if pass, useful error message otherwise

import numpy as np
def DotProduct():
  array1 = np.random.randn(5)
  array2 = np.random.randn(5)

  if isinstance(array1,np.ndarray) and isinstance(array2,np.ndarray) and array1.size == array2.size:
    print ("Os elementos so arrays e tem o mesmo tamanho")

    array3 = array1*array2
    result = array3.sum()
    return(float(result))

  elif isinstance(array1,np.ndarray) and isinstance(array2,np.ndarray) and array1.size != array2.size:
    raise Exception("Os elementos so arrays mas no tem o mesmo tamanho. Corrija.")
  elif not isinstance(array1,np.ndarray) and isinstance(array2,np.ndarray) and len(array1) == len(array2):
    raise Exception("Nem todos so arrays, mas tem o mesmo tamanho. Corrija.")
  else:
    raise Exception("Nem todos so arrays e no tem o mesmo tamanho. Corrija.")

DotProduct()

```

```python
# exercise: make poisson-like counter

# lambda = 10
# start at 1 and multiply it by random number uniform ([0,1]) and then keep multiplying the result by other random numbers
# do this until the obtained value is smaller than exp(-lambda)
# count number of iterations until obtained value < exp(-lambda)

import numpy as np
import math
lambda_value = 10
numbers = [1]

while numbers[-1] >= math.exp(-lambda_value):
  numbers.append(numbers[-1]*np.random.uniform(0,1))
  iterationcount = len(numbers)-1

print(numbers)
print(iterationcount)
```

```python
#exercise:
# 1) create a list of integers from 0 to 8(v)
# 2) reshape to a 3x3 matrix (M)
# 3) repeat that matrix to 9x3(C)
# 4) broadcast multiply with initial vector (B)

import numpy as np

v = [0,1,2,3,4,5,6,7,8]

M = np.array(v).reshape(3,3)

C = np.tile(M,(3,1))

B = (np.array(v).reshape(9,1))*C

print(v),print(" ")
print(M),print(" ");
print(C),print(" ");
print(B),print(" ");
```
